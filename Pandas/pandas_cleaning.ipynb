```python
import pandas as pd
from datasets import load_dataset
```

```python
# Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
df.info()
```

Transform the job_posted_date from a string object to datetime object

```python
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df.job_posted_date.dt.month
```

##  Add a new column, job_posted_month, to the dataset

```python
df["job_posted_month"] = df.job_posted_date.dt.month
```

```python
df.info()
```

## Sorting values using .sort_values() function

```python
df.sort_values(by="job_posted_date", inplace=True)
df
```

## Using daraFrane.drop to remove rows or columns

```python
df.drop(labels="salary_hour_avg", axis=1, inplace=True)
```

```python
df.info()
```

## Using .dropna to remove rows with missing values

```python
df.dropna(subset=["salary_year_avg"], inplace=True)
```
