# Matplotlib
## Practicing matplotlib library!

## Install all necessary libraries

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np
```

## Loading the data

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

## Transform the date into a DateTime object

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df["job_posted_date"]
```

## Quick example from the Matplotlib library

```python
x = np.arange(0,5, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

## Another simple example

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x,y)
plt.show()
```

## Plotting Job Posting Dates

```python
date_counts = df.job_posted_date.value_counts()
date_counts
```

```python
plt.plot(date_counts.index, date_counts)
```

## The graph is a hot mess! How can we fix it?

```python
date_counts = date_counts.sort_index()
```

```python
plt.plot(date_counts.index, date_counts)
```

## Good, but not good enough. 

```python
df["job_posted_month"] = df["job_posted_date"].dt.month
df
```

```python
monthly_counts = df.job_posted_month.value_counts()

monthly_counts = monthly_counts.sort_index()

monthly_counts
```

## Let's try one more time  

```python
plt.plot(monthly_counts.index, monthly_counts)
```
