# Medium Exercises

1. Divisibility Check: Given a number 12, print "Divisible by 3" if it is divisible by 3, otherwise print "Not divisible by 3".

```python
number = int(input("Enter a number: "))

if number % 3 == 0:
    print(f"{number} is divisible by 3")
else:
    print(f"{number} is not divisible by 3")
```

2. Nested Conditionals: Given a number 15, print "High" if the number is greater than 10 and "Low" if it is 10 or less.

```python
number = int(input("Enter a number: "))

if number > 10:
    print(f"{number} is greater than 10")
elif number == 10:
    print(f"{number} is equal 10")
else:
    print(f"{number} is less than 10")
```

3. String Case: Given the string "Python", print "Uppercase" if the string is uppercase, "Lowercase" if it is lowercase, and "Mixed case" if it is neither.


```python
str_input = input("Enter a string: ")

if any(str_input.isupper() for char in str_input) and any(str_input.islower() for char in str_input):
    print(f"{str_input} is an mixed case string")
elif str_input.isupper():
    print(f"{str_input} is an uppercase case string")
else:
    print(f"{str_input} is an lowercase string")
```
