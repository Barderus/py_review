3. String Case: Given the string "Python", print "Uppercase" if the string is uppercase, "Lowercase" if it is lowercase, and "Mixed case" if it is neither.


```python
str_input = input("Enter a string: ")

if any(char.isupper() for char in str_input) and any(char.islower() for char in str_input):
    print(f"{str_input} is an mixed case string")
elif str_input.isupper():
    print(f"{str_input} is an uppercase case string")
else:
    print(f"{str_input} is an lowercase string")
```

7. Leap Year: Given a year 2024, print "Leap year" if it is a leap year, otherwise print "Not a leap year". (A year is a leap year if divisible by 4, but not divisible by 100, unless also divisible by 400.)

```python
year = int(input("Enter a year: "))

if year % 400 == 0 and year % 100 == 0:
    print(f"{year} is a leap year")
elif year % 4 == 0 and year % 100 != 0:
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

1. Discount Calculation: Given a purchase amount 150, apply a 10% discount if the amount is greater than 100, otherwise apply a 5% discount. Print the final amount after discount.


```python
purchase = int(input("Enter purchase amount: "))

if purchase < 150:
    purchase += purchase * 0.05
else:
    purchase += purchase * 0.1
    
print(f"Total to spent: ${purchase}")
```

2. Triangle Classification: Given the sides of a triangle a = 3, b = 4, and c = 5, classify the triangle as "Equilateral", "Isosceles", or "Scalene" based on the lengths of the sides.

```python
a = int(input("Enter a length of side a: "))
b = int(input("Enter a length of side b: "))
c = int(input("Enter a length of side c: "))

# Equileteral = a == b == c
# Isosceles = a == b and != c
# Scalebe = a != b != c

if a == b and a == c and a == b:
    print("Equilateral triangle")
elif (a == b and b != c) or (a != b and b == c ) or (b != a and c == a) :
    print("Isosceles triangle")
else:
    print("Scalene")    
```

3. BMI Calculation: Given weight 70 kg and height 1.75 meters, calculate the BMI and print "Underweight", "Normal weight", "Overweight", or "Obesity" based on BMI ranges.


```python
height = float(input("Enter a height (in meters): "))
weight = float(input("Enter a weight (in kilograms): "))

#BMI == weight / height^2
bmi = weight / (height * height)

if bmi <= 18.5:
    print("Underweight")
elif bmi <= 25:
    print("Normal weight")
elif bmi < 40:
    print("Overweight")
else:
    print("Obesity")
```
